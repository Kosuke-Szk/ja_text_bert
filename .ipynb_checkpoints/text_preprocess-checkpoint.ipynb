複数のtxtファイルをひとつに結合する

```python
import glob
import os

f=open('./tmp.txt','w')
for directory in glob.glob('./extracted/*'):
    for name in glob.glob(directory+'/*'):
        for l in open(name,'r'):  
            list2 = l
            f.write(list2)
f.close()
```
